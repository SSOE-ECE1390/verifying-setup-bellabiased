**IMPORTS** - [O]

```python
import cv2

import os
import numpy as np

import matplotlib.pyplot as plt
```

**LOAD A GRAYSCALE IMAGE** - [O]

```python
apollo = cv2.imread(os.path.relpath('Data/Apollo_11_Launch.jpg'))
```

**VIEW IMAGE** - [O]

```python
plt.imshow(apollo);
# print(apollo.shape)
```

**CROP IMAGE** - [O]

```python
# crop image []
cropollo = apollo[420:690, 420:690]

plt.imshow(cropollo);
```

**EDIT BRIGHTNESS OF IMAGE** - [O]

```python
matrix = np.ones(cropollo.shape, dtype="uint8") * 200
brightpollo = cv2.add(cropollo, matrix)

plt.imshow(brightpollo);
# AHAHAHAHAH
# the gateway to heaven...
```

**ROTATE IMAGE** - [O]

```python
rotpollo = np.rot90(brightpollo)

plt.imshow(rotpollo);
```

**SAVE MODIFIED IMAGE** - [O]

```python
finalpollo = rotpollo;
# having way too much fun with this

cv2.imwrite(os.path.relpath('Data/Modified_Apollo_11_Launch.jpg'), finalpollo)

check = cv2.imread(os.path.relpath('Data/Modified_Apollo_11_Launch.jpg'))
plt.imshow(check);
```
